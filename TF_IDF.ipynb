```python
def get_term_frequency(input_text):
  wordlist = re.findall(r"[\w']+" , input_text)
  clean_wordlist = [w.lower() for w in wordlist if w.lower() not in all_stopwords]
  cleanwordcounts = Counter(clean_wordlist)
  cleanwordcounts = dict(cleanwordcounts)
  #total_count == .........
  total_count = len(wordlist)
  normalized_cleanwordounts = {k : v / total_count for k,v in cleanwordcounts.items()}
  return normalized_cleanwordounts
```

```python
testword = "gotham"

document_counter = 0
for index, row in df.iterrows():
  if testword in list(row["TF"].keys()):
    document_counter += 1
    print("Movie Name", row["Movie Name"]," ", row["Wiki link"])
print("\nDocument Counter: ", document_counter)
print("\nTotal Number of Movies", df.shape[0])
```

```python
N = df.shape[0]
document_frequency_of_a_word = document_counter/N
print(document_frequency_of_a_word)
```

```python
IDF_of_a_word = N/document_counter
print(IDF_of_a_word)
```

```python
#Lets avoid dividing by zero, we add an extra one in the numerator and the denominator
IDF_of_a_word = (N+1)/(document_counter + 1)
print(IDF_of_a_word)
```

```python
import math
IDF = math.log10(IDF_of_a_word)+ 1
print(IDF)
```

```python
def get_inverse_document_frequency(word, full_dataframe):
  wrd = word.lower()
  document_counter = 0
  for index, row in df.iterrows():
    if testword in list(row["TF"].keys()):
      document_counter += 1

  total_number_of_documents = full_dataframe.shape[0]
  IDF_of_word = (1 + total_number_of_documents)/(document_counter + 1)
  IDF = math.log10(IDF_of_word) + 1
  return IDF
```

### The single algorithm of the TFIDF can be used to:
1. Extract Keywords from any given document.
2. Retrieve similar documents(recommendation engine).
3. Summarize documents.
4. Sentiment Analysis.
