```python
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_trf")
```

```python
doc = nlp(batmanbegins_plot)
```

```python
displacy.render(doc,style='ent',jupyter=True)
```

```python
persons =[]
locations=[]

for token in doc.ents:
    if token.label_ =='PERSON':
      persons.append(token.text)
    if token.label_ =='GPE':
      locations.append(token.text)

persons = list(set(persons))
locations = list(set(locations))
print(persons)
print(locations)
```

```python
import gensim.downloader as api
model = api.load("word2vec-google-news-300")
```

```python
# Idli -India+ USA = Blueberry Cheesecake
# (Idli -India) = breakfast item or snack
model.most_similar(positive=['USA', 'Idli'],negative=['India'],topn=10)
```

```python
Origin_country='USA'
Target_country='India'


words=[]
similarwordslistgroup=[]

for word in persons:
  word = word.strip()
  word = word.replace(" ","_")
  try:
    similar_words_list= model.most_similar(positive=[Target_country,word],negative=[Origin_country],topn=10)
    # Remove the scores for the retrieved choices
    similar_words_list = [choices[0].replace("_"," ") for choices in similar_words_list ]
  except:
    similar_words_list = []
    # print (" Fetching similar words failed for ",word)
  print (word," -- Replacement suggestions --> ",similar_words_list)
  if len(similar_words_list)>0:
    words.append(word)
    similarwordslistgroup.append(similar_words_list)

  print ("\n")
```

```python
import ipywidgets as widgets                        # Creating widgets
from IPython.display import display                 # Displaying widgets

output = widgets.Output()
# Creating dropdown objects
# key = "Batman"
# opts = ['Caped Crusader', 'Feluda', 'Krissh', 'superhero', 'Amitabh Bachchan', 'Krrish', 'Hanuman', 'Batman Begins', 'superstar Rajinikanth', 'Sholay']
dropdownobjects = []
for key,opts in zip(words,similarwordslistgroup):
  dropdownobjects.append(widgets.Dropdown(options = opts, description=key))

# dropdownobjects =[widgets.Dropdown(options = opts, description=key),widgets.Dropdown(options = opts, description=key)]

# Display the dropdowns
input_widgets = widgets.VBox(dropdownobjects)
print ("Choose from the dropdown the best replacement for a given word and proceed to the next cell : \n")
display(input_widgets)
```

```python
from sentence_transformers import SentenceTransformer
# all-MiniLM-L6-v2
# all-mpnet-base-v2
# https://www.sbert.net/docs/pretrained_models.html
sentence_encoder_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
df
```

```python
print (df.shape)
df.drop_duplicates(subset ="Wiki link", keep = False, inplace = True)
print (df.shape)
```

```python
movie_plots = df['Plot'].tolist()
print (len(movie_plots))
print (movie_plots[0])
```

```python
embeddings = sentence_encoder_model.encode(movie_plots,show_progress_bar=True)
```

```python
print (len(embeddings[0]))
print (embeddings[0])
```

```python
df['full_embeddings']= embeddings.tolist()
df.head(3)
```

```python
import umap
import numpy as np

# embeddings - np.array
umap_embeddings_2dim = umap.UMAP(n_neighbors=5,
                            n_components=2,
                            min_dist=0.0,metric='cosine').fit_transform(embeddings)

df["umap_embeddings_2dim"]  = umap_embeddings_2dim.tolist()
df.head(5)
```

```python
import numpy as np
import plotly.express as px

col="umap_embeddings_2dim"

hover_data =["Wiki link"]
hover_name = "Movie Name"
plot_values = np.stack(df[col], axis=1)


x, y = plot_values[0], plot_values[1]
fig = px.scatter(
            df,
            x=x,
            y=y,
            hover_data=hover_data,
            title="Visualize Movie Scripts",
            hover_name=hover_name,
        )

fig.show()
```

### This data was collected related to movie directors who were really influential in Hollywood

```python
directors_desc_path = "/content/drive/My Drive/NLP/top_100_directors_description.csv"

import pandas as pd
df_dirs = pd.read_csv(directors_desc_path)
print (df_dirs.shape)
df_dirs.head(5)
```

```python
dir_bios = df_dirs['selection1_Biography'].tolist()
print (dir_bios[0])
```

```python
embeddings_dirs = sentence_encoder_model.encode(dir_bios,show_progress_bar=True)
df_dirs['full_embeddings']= embeddings_dirs.tolist()
df_dirs.head()
```

```python
df_dirs['type']='Director'
df_dirs.head()
```

```python
df_directors = df_dirs.rename(columns={'selection1_name': 'Name', 'selection1_url': 'Link'})
df_directors =   df_directors[["Name", "Link","full_embeddings","type"]]
df_directors.head()
```

```python
df_movies = df.rename(columns={'Movie Name': 'Name', 'Wiki link': 'Link'})
df_movies =   df_movies[["Name", "Link","full_embeddings","type"]]
df_movies.head(3)
```

```python
combined_df = [df_movies, df_directors]
combined_df_result = pd.concat(combined_df)
combined_df_result.sample()
```

```python
full_embeddings_list = np.array(combined_df_result['full_embeddings'].tolist())
```

```python
umap_embeddings_2dim_combined = umap.UMAP(n_neighbors=5,
                            n_components=2,
                            min_dist=0.0,metric='cosine').fit_transform(full_embeddings_list)
```

```python
combined_df_result["umap_embeddings_2dim"]  = umap_embeddings_2dim_combined.tolist()
combined_df_result.head()
```

```python
import numpy as np
import plotly.express as px

col="umap_embeddings_2dim"

hover_data =["Link"]
hover_name = "Name"
plot_values = np.stack(combined_df_result[col], axis=1)
color = "type"

x, y = plot_values[0], plot_values[1]
fig = px.scatter(
            combined_df_result,
            x=x,
            y=y,
            color = color,
            hover_data=hover_data,
            title="Visualize Movie Scripts and Directors",
            hover_name=hover_name
        )

fig.show()
```
